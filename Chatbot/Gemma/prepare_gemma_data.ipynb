```python
import pandas as pd
import json
from sklearn.model_selection import train_test_split

def format_gemma_chat(user, assistant):
    user = user.strip()
    assistant = assistant.strip()
    return f"<bos><start_of_turn>user\n{user}<end_of_turn>\n<start_of_turn>model\n{assistant}<end_of_turn><eos>"

df = pd.read_csv('gemma_finetune_movie_mood_dataset_v3_with_edge_cases.csv')

df = df.dropna(subset=['user', 'assistant'])

df['text'] = df.apply(lambda row: format_gemma_chat(row['user'], row['assistant']), axis=1)

# Tách train/test (80/20)
train_df, test_df = train_test_split(df, test_size=0.2, random_state=42)

# Lưu thành file JSONL (mỗi dòng là một JSON)
train_df[['text']].to_json('train.jsonl', orient='records', lines=True)
test_df[['text']].to_json('test.jsonl', orient='records', lines=True)

print(f"Tổng số mẫu: {len(df)}")
print(f"Train: {len(train_df)}, Test: {len(test_df)}")
```
